```python
import pygame
import string
import time

# External autofill function
def external_autofill_function(selected_letters):
  return "EXAMPLE"

# Simulate external P300 classification
def external_p300_classify(): 
  time.sleep(2) # Simulate delay
  return True # or False
```

```python
# Initialize PyGame 
pygame.init()

# Display keyboard 
keyboard = list(string.ascii_lowercase + 'Space') 

# Set up screen
screen = pygame.display.set_mode((600,400))

# Colors
WHITE = (255,255,255)
RED = (255,0,0)
BLACK = (0,0,0)

# Load font
font = pygame.font.SysFont("comicsansms", 20)

# Create keyboard layout
keys = []
for row in range(3):
  for col in range(10):
    try:
        letter = keyboard[row*10 + col]
        keys.append([letter, (30 + col*50, 100 + row*50)])
    except:
        pass

selected_letters = []

# Main loop
while True:

  for event in pygame.event.get(): # You need pygame event handling to prevent from freezing
      if event.type == pygame.QUIT: # If user wants to quit
          pygame.quit()
  
  # Get autofill word
  autofill_word = external_autofill_function(selected_letters)
  print("autofill: "+autofill_word)

  # Set current letters
  current_letters = list(string.ascii_lowercase) + ['Space'] + [autofill_word]
  # current_letters = current_letters[:len(current_letters)/2]
  
  # Initialize current selection
  current_selection = current_letters

  # Display keyboard
  screen.fill(WHITE)
  for key in keys:
    text_surface = font.render(key[0], True, BLACK)
    screen.blit(text_surface, key[1])

  # Enable display
  time.sleep(0.5)
  pygame.display.flip()

  getLetter = True

  while getLetter:

    for event in pygame.event.get(): # You need pygame event handling to prevent from freezing
      if event.type == pygame.QUIT: # If user wants to quit
        pygame.quit()
          
    # Highlight current selection
    for key in current_selection:
      if key in keyboard:
        i = keyboard.index(key)
        pygame.draw.rect(screen, RED, pygame.Rect(keys[i][1][0]-5, keys[i][1][1]-5, 40, 40), 2)
      else:
        # Draw box for autofill word
        pygame.draw.rect(screen, RED, pygame.Rect(50, 300, 100, 40), 2)
    print("Highlight:", current_selection)
    time.sleep(0.5)
    pygame.display.flip()

    # Simulate P300 classification
    p300_response = external_p300_classify()

    # Wait a half second and then unhighlight the keys
    for key in current_selection:
      if key in keyboard:
        i = keyboard.index(key)
        pygame.draw.rect(screen, BLACK, pygame.Rect(keys[i][1][0]-5, keys[i][1][1]-5, 40, 40), 2)
      else:
        # Draw box for autofill word
        pygame.draw.rect(screen, BLACK, pygame.Rect(50, 300, 100, 40), 2)
    time.sleep(0.5)
    pygame.display.flip()
    
    # Halve selection based on response
    if p300_response:
      current_selection = current_selection[:len(current_selection)//2] 
      print(len(current_selection))
    else:
      current_selection = current_selection[len(current_selection)//2:]
      print(len(current_selection))
      print("else")
    
    # Check if only 1 letter left
    if len(current_selection) == 1:
      selected_letter = current_selection[0]
      selected_letters.append(selected_letter)
      print("Selected:", selected_letter)
      time.sleep(10)
      getLetter = False
      
  # Reset   
  print("Reset")
  time.sleep(1)
  break

pygame.quit()
```
